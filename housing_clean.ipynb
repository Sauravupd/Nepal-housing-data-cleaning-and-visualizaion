```python
def get_unit(x):
    unit = x.split()
    return unit[1]
def get_area(x):
    area = x.split()
    return area[0]
```

```python
df['Unit']=df['Area'].apply(lambda x:  f"{get_unit(x)}")
df['New_Area'] = df['Area'].apply(lambda x: f"{get_area(x)}")
```

```python
df_filtered = df[(df['Unit']!='Aana') & (df['Unit']!='Ropani')].index
df_filtered
```

```python
df.drop(df_filtered,axis=0,inplace = True)
```

```python
df2=df[~df['New_Area'].str.contains("-")]
```

```python
def get_width(x):
    z=x.split()
    return z[0]
```

```python
df2['Road_Width']=df2['Road Width'].apply(lambda x: f"{get_width(x)}")
```

```python
df3 = df2.drop(["Title","Address","Floors","Parking","Year","Views","Area","Road Width","Road Type","Build Area","Posted"],axis = 1)
```

```python
df3
```

```python
df3["Amenities"]=df3["Amenities"].str.strip('[]').str.split(',')
```

```python
df3["Number of Amenities"]=df3["Amenities"].apply(lambda x: len(x))
```

```python
df3['New_Area'] = df3['New_Area'].apply(lambda x:".".join(x.split(".")[0:2]))

```

```python
df4=df3[df3['New_Area'].str.contains('^(?:\d{1,2})?(?:\.\d{1,2})?$',regex=True)]
```

```python
# df4['New_Area']=df4['New_Area'].astype(float)
df4['New_Area']=pd.to_numeric(df4['New_Area'])
```

```python
df4['Road_Width']=df4['Road_Width'].astype(int)
```
